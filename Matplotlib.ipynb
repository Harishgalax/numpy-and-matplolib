```python
import matplotlib.pyplot as plt
%matplotlib inline
x=[1,2,3,4,5,6,7]
y=[23,35,46,57,68,88,79]
plt.plot(x,y)
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
```

```python
plt.plot(x,y,'g+')
```

```python
plt.plot(x,y,color='#0000ff',marker='D',linestyle='',markersize=20)
```

```python
plt.plot(x,y,color='green',alpha=0.5)
#alpha decides shadeness of line
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46 ]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
#many graphs together
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46 ]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
#many graphs together
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label='avg')
plt.legend(loc='best')#to localise it to the upper corner
```

```python
import numpy as np
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos=np.arange(len(company))#because company names are string, we have to convert them as array
xpos
```

```python
plt.bar(company,revenue)
plt.xlabel('Company')
plt.ylabel('Stocks')
plt.title('Stats')
plt.bar(company,profit)
```

```python
plt.bar(xpos-0.2,revenue,width=0.4)
plt.bar(xpos+0.2,profit,width=0.4) # To separate the both bars
plt.xticks(xpos,company)
plt.xlabel('Company')
plt.ylabel('Stocks')
plt.title('Stats')
```
